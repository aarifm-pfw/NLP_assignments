```python
#Importing necessary libraries
from sklearn.model_selection import train_test_split
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score
from transformers import pipeline
import numpy as np
import pandas as pd
from sklearn.linear_model import LogisticRegression
```

```python
def evaluate_model(y_true, y_pred, model_name):
    accuracy = accuracy_score(y_true, y_pred)
    precision = precision_score(y_true, y_pred, average='weighted', zero_division=0)
    recall = recall_score(y_true, y_pred, average='weighted', zero_division=0)
    f1 = f1_score(y_true, y_pred, average='weighted', zero_division=0)

    print(f"Model: {model_name}")
    print("Accuracy:", accuracy)
    print("Precision:", precision)
    print("Recall:", recall)
    print("F1 Score:", f1)
    print("-" * 40)
```

```python
data = pd.read_csv('/content/all-data.csv', encoding='ISO-8859-1')
data.columns = ['Sentiment', 'Text']
X_train, X_test, y_train, y_test = train_test_split(data['Text'], data['Sentiment'], test_size=0.2, random_state=42)
```

```python
def run_sentiment_analysis(model_name, X_test):
    sentiment_pipeline = pipeline("text-classification", model=model_name)
    predictions = sentiment_pipeline(X_test.tolist())
    return [pred['label'] for pred in predictions]
```

```python
def run_zero_shot_classification(X_test, candidate_labels=["positive", "neutral", "negative"]):
    zero_shot_pipeline = pipeline("zero-shot-classification", model="tasksource/deberta-small-long-nli")
    predictions = zero_shot_pipeline(X_test.tolist(), candidate_labels=candidate_labels, truncation=True, max_length=512)
    return [pred['labels'][0] for pred in predictions]
```

```python
y_pred_1 = run_sentiment_analysis("ahmedrachid/FinancialBERT-Sentiment-Analysis", X_test)
evaluate_model(y_test, y_pred_1, "FinancialBERT-Sentiment-Analysis")

y_pred_2 = run_sentiment_analysis("mrm8488/distilroberta-finetuned-financial-news-sentiment-analysis", X_test)
evaluate_model(y_test, y_pred_2, "distilroberta-finetuned-financial-news-sentiment-analysis")

y_pred_3 = run_zero_shot_classification(X_test)
evaluate_model(y_test, y_pred_3, "Zero-Shot Classification with DeBERTa")
```

```python
def extract_features(data):
    feature_extractor = pipeline("feature-extraction", model="Agita/DistilBERT_test")
    return [np.mean(feature_extractor(text)[0], axis=0) for text in data]

X_features = extract_features(data['Text'])
y_labels = data['Sentiment'].map({'positive': 2, 'neutral': 1, 'negative': 0})

X_train_features, X_test_features, y_train_features, y_test_features = train_test_split(X_features, y_labels, test_size=0.2, random_state=42)
```

```python
clf = LogisticRegression(solver='saga', max_iter=500)
clf.fit(X_train_features, y_train_features)

# Predict and evaluate Logistic Regression model
y_pred_lr = clf.predict(X_test_features)
evaluate_model(y_test_features, y_pred_lr, "Logistic Regression (DistilBERT Features)")
```
